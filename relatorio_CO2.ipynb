# Análise Comparativa das Emissões e Políticas Climáticas Globais

## Reino Unido
reinounido.png
- **Redução contínua de emissões nas últimas décadas.**  
- **Transição energética:** Do carvão para renováveis, como energia eólica e solar, impulsionada pela *Lei de Mudança Climática*.  
- **Políticas fiscais:** O imposto sobre carbono forçou a substituição de combustíveis fósseis.  
- **Resultado:** O Reino Unido lidera globalmente na queda das emissões, com avanços consistentes desde os anos 2000.

## China
china.png
- **Maior emissor global de CO₂,** com crescimento motivado pela industrialização rápida e urbanização.  
- **Dependência do carvão** para geração de energia.  
- Embora a China invista em energias renováveis e prometa **neutralidade até 2060**, a transição é lenta.  
- **Desafio:** A necessidade de crescimento econômico dificulta a redução substancial das emissões.

## Estados Unidos
eua.png
- **Segundo maior emissor mundial,** com uma matriz baseada em petróleo e gás natural.  
- **Inconsistências nas políticas governamentais** causam incerteza na trajetória futura das emissões.  
- **Redução do uso de carvão:** Substituído em parte por gás natural, mas este ainda é um combustível emissor de carbono.

## Rússia
russia.png
- **Emissões ligadas à queima de gás natural e petróleo**, usados para consumo interno e exportação.  
- **Dependência da exportação de combustíveis fósseis** dificulta a transição para energias renováveis.  
- **Demanda de gás natural:** A necessidade de aquecimento em climas extremos torna a mudança mais desafiadora.

## Alemanha
alemanha.png
- **Líder em energias renováveis na Europa**, com foco na transição energética (*Energiewende*).  
- **Dependência do carvão:** Ainda usado em setores industriais.  
- Apesar do progresso na energia doméstica, **políticas para a indústria** são necessárias para acelerar a redução das emissões.

---

## Impacto das Políticas Públicas
- **Reino Unido:** O fechamento de usinas a carvão e a adoção de fontes renováveis foram fundamentais para a queda nas emissões.  
- **Alemanha:** O plano *Energiewende* impulsionou a transição para energias limpas, mas desafios permanecem no setor industrial.  
- **Estados Unidos:** Políticas inconsistentes entre governos federais e estaduais limitam o progresso. A substituição do carvão por gás natural trouxe uma redução modesta nas emissões.  
- **Brasil:** Apesar de ter uma matriz energética limpa baseada em hidrelétricas, a dependência de combustíveis fósseis no transporte prejudica a redução de emissões. Políticas públicas ainda são pouco efetivas.  
- **China:** Embora haja investimentos em renováveis, a dependência do carvão impede uma redução rápida das emissões.

---

## Conclusão
A análise revela que **Reino Unido e Alemanha** são exemplos de sucesso na redução de emissões, graças a políticas públicas eficazes e investimentos em energias renováveis.  
Por outro lado, **Estados Unidos, Rússia e China** enfrentam dificuldades devido à dependência de combustíveis fósseis e ao desafio de equilibrar crescimento econômico com sustentabilidade.

- **China e Brasil:** Precisam melhorar a implementação de políticas públicas voltadas para a transição energética e o transporte.  
- **Estados Unidos e Rússia:** Enfrentam obstáculos estruturais e políticos para obter reduções mais significativas.

A análise destaca que **políticas adaptadas às realidades locais** são essenciais para que cada país contribua de forma eficaz no combate às mudanças climáticas.
